```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

TIỀN XỬ LÝ DỮ LIỆU

```python
import re

def read_and_process(filename):
  """
  Đọc file -> xử lý dữ liệu -> trả về kq (str)
  Returns:
    str: Dữ liệu đã được xử lý và nối chuỗi với ký tự xuống dòng
  """
  with open(filename, "r", encoding="utf-8") as f:
    data = f.read()

  processed_data = ""

  # Xử lý dữ liệu theo từng dòng
  for line in data.splitlines():

    # # Loại bỏ các link 
    # url_pattern = re.compile(r'http\S*')
    # line = url_pattern.sub(r'', line)

    # # Loại bỏ email
    # email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
    # line = email_pattern.sub(r'', line)

    # # Loại bỏ sdt
    # phone_pattern = re.compile(r'\b\d{10,15}\b|\b\d{2,4}[-.\s]?\d{2,4}[-.\s]?\d{3,4}\b|\b\d{4}\s\d{3}\s\d{3}\b|\+\d{1,4}[-.\s]?\d{1,4}[-.\s]?\d{1,4}[-.\s]?\d{1,9}\b|\b\d{3}[-.\s]?\d{3,4}[-.\s]?\d{3,4}\b|\b\d{8,9}\b|\b\d{4}\s\d{2}\s\d{2}\s\d{2}\b')
    # line = phone_pattern.sub(r'', line)
  
    # Loại bỏ ký tự đặc biệt
    line = re.sub(r"[^\w\s]", "", line)

    # Loại bỏ khoảng trắng thừa
    pattern = r"\s+"
    line = re.sub(pattern, " ", line)

    # Chuyển đổi sang chữ thường
    line = line.lower()

    # Lưu dữ liệu đã được xử lý và nối chuỗi với kt xuống dòng
    processed_data += line + "\n"

  # Loại bỏ ký tự xuống dòng thừa ở cuối
  processed_data = processed_data[:-1]

  return processed_data

filename_train = "3.topic_detection_train.v1.0.txt"
filename_test = "3.topic_detection_test.v1.0.txt"

# Đọc và xử lý dữ liệu
processed_data_train = read_and_process(filename_train)
processed_data_test =read_and_process(filename_test)

# print(processed_data_train)
# print(processed_data_test)
```

CHUẨN BỊ DỮ LIỆU

```python
# Tim kiếm tất cả các cặp nhãn và nội dung trong tập đã xử lý và đưa chúng vào tuple
pattern = r'(__label__\w+)\s((?:.(?!__label__))+)' 
matches_train = re.findall(pattern, processed_data_train, re.DOTALL)
matches_test = re.findall(pattern, processed_data_test, re.DOTALL)

# for match in matches_train:
#     print(match)

#Chuyển đổi danh sách các nhãn và nội dung thành DataFrame
train_df = pd.DataFrame(matches_train, columns=['label', 'content'])
test_df = pd.DataFrame(matches_test, columns=['label', 'content'])
pd.set_option('display.max_colwidth', None)

# kiểm tra
test_df.head(6)
```

```python
label_counts = train_df['label'].value_counts()
print("Số lượng mẫu cho từng nhãn:")
print(label_counts)
```

```python

with open('vietnamese-stopwords.txt', 'r', encoding='utf-8') as f:
    vietnamese_stopwords = [line.strip() for line in f]

## Biểu diễn văn bản bằng TF-IDF
tfidf = TfidfVectorizer(stop_words=vietnamese_stopwords,max_features=1500)
X_train_tfidf = tfidf.fit_transform(train_df['content'])
X_test_tfidf = tfidf.transform(test_df['content'])
X_test_tfidf
```

```python
# Biểu diễn nhãn dưới dạng số
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(train_df['label'])
y_test = le.transform(test_df['label'])
y_test
```

```python
from scipy.sparse import hstack
# Bổ sung thêm đặc trưng (độ dài văn bản)
train_text_lengths = [len(doc.split()) for doc in train_df['content']]
test_text_lengths = [len(doc.split()) for doc in test_df['content']]
```

```python
import numpy as np
X_train_other_features = hstack((X_train_tfidf, np.array(train_text_lengths)[:, None]))
X_test_other_features = hstack((X_test_tfidf, np.array(test_text_lengths)[:, None]))
```

HUẤN LUYỆN MÔ HÌNH

```python
# Huấn luyện mô hình Logistic Regression với TF-IDF(dùng thuật toán liblinear)
model = LogisticRegression(solver='liblinear', max_iter=150)
model.fit(X_train_other_features, y_train)
```

```python
# In ra thuật toán được sử dụng
print(f"Solver used: {model.solver}")
```

ĐÁNH GIÁ MÔ HÌNH

```python
# Đánh giá mô hình
y_pred = model.predict(X_test_other_features)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))
```

CHECK

```python
for i in range(10):  # in ra 10 dự đoán đầu tiên
    index = i  # chỉ số của mẫu trong tập test
    predicted_label = le.inverse_transform([y_pred[index]])[0]
    true_label = test_df.loc[index, 'label']
    text = test_df.loc[index, 'content']
    print(f"content: {text[:100]}...")
    print(f"True Label: {true_label}, Predicted Label: {predicted_label}")
    print()
```

```python
from sklearn.metrics import accuracy_score

# Độ chính xác trên tập test
accuracy = accuracy_score(y_test, y_pred)
print(f"Độ chính xác của mô hình Logistic Regression trên tập test: {accuracy:.10f}")
```

```python
from sklearn.svm import SVC

# Huấn luyện mô hình SVM với TF-IDF
model_svm = SVC(kernel='linear', random_state=42)
model_svm.fit(X_train_tfidf, y_train)
```

```python
# Đánh giá mô hình SVM trên tập test
y_pred_svm = model_svm.predict(X_test_tfidf)
print("SVM with TF-IDF Classification Report:\n", classification_report(y_test, y_pred_svm, target_names=le.classes_))
```

```python
from sklearn.metrics import accuracy_score

# Độ chính xác mô hình SVM trên tập test
accuracy = accuracy_score(y_test, y_pred_svm)
print(f"Độ chính xác của SVM trên tập test: {accuracy:.10f}")
```
